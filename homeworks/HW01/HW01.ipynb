```python
import numpy as np

array = [1,2,3,4,5]
array2D = ([[1,2,3], [2,3,4], [3,4,5]]) #двумерный массив

print(array)
print(array2D)
```

```python
array = np.arange(9)
array = array.reshape(3,3) #изменение массива в матрицу 3на3
print(array)
```

```python
zeros = np.zeros(6) #массив из нулей длинной в 6 элементов
ones = np.ones((4,4)) #матрица 4на4 из единиц
array = np.arange(-10, 10, 2).reshape(2,5) #arange(нач точка,конечная точка,шаг)
zeros, ones, array
```

```python
A = np.array([[1, 2],[5, 6]])
B = np.array([[-2, 6],[4, -5]])
print(A*B)
```

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
print("A:\n", A)
print("\nTranspose A:\n", np.transpose(A))
```
